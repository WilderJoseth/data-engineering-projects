# Pre bronze layer

```python
in_parameter_id_run = 1
in_parameter_process_date = "1900-01-01 00:00:00"
```

```python
path_root = "Files/raw"
path_invalid = "Files/invalid"
expected_columns = "Company,Location,Date,Time,Rocket,Mission,RocketStatus,Price,MissionStatus"
```

```python
file_columns_read = []

files = mssparkutils.fs.ls(path_root)
for f in files:
    df = spark.read.text(f"{path_root}/{f.name}")
    header_line = df.first()[0]
    file_columns_read.append((f.name, header_line))
```

```python
file_results = []
for f, c in file_columns_read:
    try:
        if expected_columns == c:
            file_results.append((f, 1))
        else:
            file_results.append((f, 0))
    except:
        file_results.append((f, 0))
```

```python
id_run_str = str(in_parameter_id_run)
process_date_str = in_parameter_process_date.replace(" ", "-").replace(":", "-")
i = 1
for f, is_valid in file_results:
    if is_valid == 0:
        i_str = str(i)
        mssparkutils.fs.mv(f"{path_root}/{f}", f"{path_invalid}/{i_str}-{id_run_str}-{process_date_str}.csv")
        i += 1
```
