# Bronze layer

```python
in_parameter_id_run = 1
in_parameter_process_date = "1900-01-01 00:00:00"
out_parameter_count_processed = 0
```

## 1. Create objects

```python
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

## 2. Extract data

```python
mission_schema = StructType([
    StructField("company", StringType()),
    StructField("location", StringType()),
    StructField("date", StringType()),
    StructField("time", StringType()),
    StructField("rocket", StringType()),
    StructField("mission", StringType()),
    StructField("rocket_status", StringType()),
    StructField("price", StringType()),
    StructField("mission_status", StringType())
    ])

df = spark.read.load("Files/raw/*.csv", schema = mission_schema, format = "csv", header = True)
```

Add run identifier

```python
df = df.withColumn("process_datetime", lit(in_parameter_process_date))
df = df.withColumn("process_datetime", to_timestamp("process_datetime", "yyyy-MM-dd HH:mm:ss"))
df = df.withColumn("process_date", to_date("process_datetime", "yyyy-MM-dd"))
df = df.withColumn("run_id", lit(in_parameter_id_run))
```

## 3. Load data

```python
df.write.partitionBy("process_date", "run_id").format("delta").mode("append").saveAsTable("bronze.missions")
```

```python
out_parameter_count_processed = df.count()
```
