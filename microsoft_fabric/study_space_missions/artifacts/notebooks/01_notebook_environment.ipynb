## Create tables

```python
from delta.tables import *
```

### Control tables

Log

```python
DeltaTable.createIfNotExists(spark) \
    .tableName("control.logs") \
    .addColumn("run_id", "INT", True) \
    .addColumn("code", "STRING", True) \
    .addColumn("description", "STRING", True) \
  .execute()
```

Validation Rules

```python
DeltaTable.createIfNotExists(spark) \
    .tableName("control.validation_rules") \
    .addColumn("schema_name", "STRING", True) \
    .addColumn("table_name", "STRING", True) \
    .addColumn("column_order", "SHORT", True) \
    .addColumn("column_name_original", "STRING", True) \
    .addColumn("column_name_new", "STRING", True) \
    .addColumn("column_type", "STRING", True) \
    .addColumn("column_size", "SHORT", True) \
    .addColumn("column_size_scale", "SHORT", True) \
    .addColumn("default_value", "STRING", True) \
    .addColumn("is_from_file", "BOOLEAN", True) \
    .addColumn("is_active", "BOOLEAN", True) \
  .execute()
```

### Bronze table

```python
DeltaTable.createIfNotExists(spark) \
    .tableName("bronze.missions") \
    .addColumn("company", "STRING", True) \
    .addColumn("location", "STRING", True) \
    .addColumn("date", "STRING", True) \
    .addColumn("time", "STRING", True) \
    .addColumn("rocket", "STRING", True) \
    .addColumn("mission", "STRING", True) \
    .addColumn("rocket_status", "STRING", True) \
    .addColumn("price", "STRING", True) \
    .addColumn("mission_status", "STRING", True) \
    .addColumn("process_datetime", "TIMESTAMP", True) \
    .addColumn("process_date", "DATE", True) \
    .addColumn("run_id", "INT", True) \
    .partitionedBy(["process_date", "run_id"]) \
  .execute()
```

### Silver table

```python
DeltaTable.createIfNotExists(spark) \
    .tableName("silver.missions") \
    .addColumn("company", "STRING", True) \
    .addColumn("location", "STRING", True) \
    .addColumn("date", "DATE", True) \
    .addColumn("rocket", "STRING", True) \
    .addColumn("mission", "STRING", True) \
    .addColumn("rocket_status", "STRING", True) \
    .addColumn("price", "DECIMAL(10, 2)", True) \
    .addColumn("mission_status", "STRING", True) \
    .addColumn("country", "STRING", True) \
    .addColumn("process_datetime", "TIMESTAMP", True) \
    .addColumn("process_date", "DATE", True) \
    .addColumn("run_id", "INT", True) \
    .partitionedBy(["process_date", "run_id"]) \
  .execute()
```
